```python
import pandas as pd
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler
import torch
import os
import pyexiv2
```

```python
def write_prompt_in_exif(image_path,prompt):
  '''
  EXIF情報を書き込む
  '''
  with pyexiv2.Image(image_path) as img:
      img.modify_exif({'Exif.Photo.UserComment': prompt}) 
```

```python
prompt_csv = './prompt_seed.csv'
df = pd.read_csv(prompt_csv)
target_df = df[df['generated'] == 'no']
```

```python
target_df
```

```python
# prepare model
model_id = "stabilityai/stable-diffusion-2-1"

# Use the DPMSolverMultistepScheduler (DPM-Solver++) scheduler here instead
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to("cuda")
```

```python
output_dir = './output'
if not os.path.exists(output_dir):
  os.makedirs(output_dir)

GEN_IMAGES = 4

for i,prompt in enumerate(target_df.iterrows()):
  target_prompt = prompt[1][0]
  print(target_prompt)
  for idx in range(GEN_IMAGES):
    test = pipe(target_prompt)
    image = test.images[0]
    output_image = os.path.join(output_dir,'{}_{}.jpeg'.format(i,idx))
    image.save(output_image)
    write_prompt_in_exif(output_image,target_prompt)
```
